```python
import numpy as np
import matplotlib.pyplot as plt
def plot_my_vector(vecs):
    i=0
    col=['r','g','b','y']
    for vec in vecs:
        plt.quiver(vec[0],vec[1],vec[2],vec[3],scale_units='xy',angles='xy',scale=100,color=col[i%len(col)])
        i+=1
        
plt.xlim=(-1,1)
plt.ylim=(-1,1)

vecs=[np.asarray([0,0,3,4]),np.asarray([0,0,-3,1])]
plot_my_vector(vecs)

plot_my_vector(vecs)
plot_my_vector([vecs[0],vecs[1],vecs[0]+vecs[1],vecs[0]-vecs[1]])

plt.show()
```

```python
#Working with dot products
vecs2=[np.asarray([0,0,-3,4]),np.asarray([0,0,3,4])]
plot_my_vector(vecs2)
```

```python
a,b=vecs2[0],vecs2[1]
#Finding dot product of a,b
a_dot_b=np.dot(a,b)     # This gives us a scalar value


#Finding the magnitude or normalising a vector 
mag_of_b=np.linalg.norm(b)

#Finding and ploting the projection of A on B
proj_of_a_on_b=a_dot_b/(mag_of_b)
proj_of_a_on_b        #This is magnitude of the projection

vec_a_on_b=proj_of_a_on_b*b
vec_a_on_b
plot_my_vector([vec_a_on_b,])
plot_my_vector(vecs2)
```
