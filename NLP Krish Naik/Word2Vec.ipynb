# Word2Vec Vectorization

Bag of Vectors and TF-IDF Problems:
- In both Bag of Words and TF-IDF the semantic information is not stored. TF-IDF gives importance to uncommon words.
- There is a definite chance of over-fitting.

- In Word2Vec; each word is basically represented as a vector of 32 or more dimension instead of a sigle number.
- Here, the semantic information and the relation between different numbers is also preserved.
